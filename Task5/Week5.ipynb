```python
# Question 5.6

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.concatenate((a, b))) # The result is a 1 by 6 matrix (1D array)

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print(np.concatenate((a, b)))  # The result is a 4 by 2 matrix (2D array)
print(np.concatenate((a, b), axis=0))
print(np.concatenate((a, b), axis=1)) 
 
# The first two are the same,  axis = 1 makes the elements of the first column the elements of the first row
```

```python
# Question 5.10

import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('ngc6946.txt', comments="#")
newdata = data.reshape(40,50,70)  # reshaping the data



res = newdata[:, 25, 35] # printing z values around the middle of the xy plane
print(res)

# Elementary statistics
a,b,c,d,e = np.min(res), np.max(res), np.sum(res), np.mean(res), np.std(res)

print(f"min={np.min(res)}, max ={np.max(res)}, sum={np.sum(res)}, mean={np.mean(res)}, std={np.std(res)}")

plt.plot(res)
plt.text(25, 0.06, f'min = {a}')
plt.text(25, 0.055, f'max = {b}')
plt.text(25, 0.05, f'sum = {c}')
plt.text(25, 0.045, f'mean = {d}')
plt.text(20, 0.04, f'std = {e}')
plt.show()
```

```python
# Question 5.11

import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('ngc6946.txt', comments="#")
newdata = data.reshape(40,50,70)  # reshaping the data

image = newdata[0]

plt.imshow(image, origin = 'lower')
plt.colorbar()

mean = np.mean(newdata)
std = np.std(newdata)

levels = [mean, mean+std, mean+2*std]

plt.contour(image, levels = levels, colors = 'red')

plt.show()
```

```python
# Question 5.12

import numpy
import scipy.constants
from scipy.constants import c

class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      # Calcium H line in rest frame
    c = c # defining c as a class variable
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
    def redshift(self):        
        z = (self.caH - self.caH_0)/self.caH
        return z
    def velocity(self): # definiting method valocity
        v = Galaxy.redshift(self)*c
        return v
        
gal2 = Galaxy("NGC1832", 3994.7)
name = gal2.name
z = gal2.redshift()
v = gal2.velocity()
print(f"Redshift of {name} is {z}")
print(f"The velocity of {name} is {v}")
```
