```python
# Funky number guessing game
import math
import array
import numpy as np

x = np.random.randint(0, 100)

y = int(input("Guess the number!!!"))

while x!= y:
    if y > x:
        y = int(input("Go lowerrrrrr"))
    else:
        y = int(input("Go higherrrrrr"))

if y == x:
    print("Wowza you did that!")
```
