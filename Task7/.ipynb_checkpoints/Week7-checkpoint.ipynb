```python
# Question 7.1

import numpy as np
from numpy import linalg as la

A = np.array([1,2,3,0,-4,1,0,3,-1]).reshape(3,3)
print(A)

# Finding the determinant
det = la.det(A)
print(f"Determinant of A is: {det}")

# Solving the system of equations Ax = b -> x = Ainv*b
b = np.array([2,2,-1])
Ainv = la.inv(A)
xyz = Ainv@b

print(f"\nSolution: x, y, z = {xyz}")

# Checking if its orthogonal
Atrans = A.T
print(A@Atrans) # not orthogonal (the result isn't an identity matrix)

# Finding the Eigenvalues
Lambda, v = la.eig(A)
print("Eigenvalues are:", Lambda)
```

```python
import numpy as np
from numpy import linalg as la
import math as m

# Converting the angles to radians
a = m.radians(57)
b = m.radians(62.6)
c = m.radians(192.25)

# Calculating the rotaion matrices which are part of the composed rotation matrix E
Ez = np.array([np.cos(a),np.sin(a),0,-np.sin(a),np.cos(a),0,0,0,1]).reshape(3,3)
Ey = np.array([np.cos(b),0,-np.sin(b),0,1,0,np.sin(b),0,np.cos(b)]).reshape(3,3)
Ez2 = np.array([np.cos(c),np.sin(c),0,-np.sin(c),np.cos(c),0,0,0,1]).reshape(3,3)

# Calculating the composed matrix E
E = Ez@Ey@Ez2
print(E)

# Proving that its a rotation matrix, that is uts inverse is the transpose of E, and its determinant is 1

Et = E.T
Einv = la.inv(E)
det = la.det(E)
print(f"  {Et} = \n{Einv} \n and the determinant is {det} \n")

# Calculating eigenvalue

Lambda, v = la.eig(E)
print(f"eigenvalues = {Lambda}")

# Defining the vector v

alpha = m.radians(192.25)
delta = m.radians(27.4)

# Using equation 7.16 to calculate v
v = np.array([np.cos(alpha)*np.cos(delta), np.cos(delta)*np.sin(alpha), np.sin(delta)]).reshape(3,1)
# Calculating the norm of v
normv = la.norm(v)
print(f"{v} \n nomr of v is {normv}")

# Switching coordinates

v2 = E@v
print(v2) # it doesn't give zero but 10^-17 which is lowky the same so it works
```

```python
import numpy as np
from numpy import linalg as la
import math as m
import matplotlib.pyplot as plt

x = np.array([1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9])
y = np.array([0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0])

a_val = []
for i in reversed(range(len(x))): # calculating the matrix entries, reversed switched the order of the rows
    a = x**(i)
    a_val.append(a)

a_val_array = np.array(a_val) # changing the list into an array

matrix = a_val_array.reshape(len(x),len(x)) # creating the matrix
y = y.reshape(len(y),1)

A = matrix.T # transposing the matrix so that the first column becomes the first row
print(A)

coeffs = la.solve(A,y) # finding the coefficients of the polynomial, transposing the matrix to get the correct one

newx = np.linspace(np.min(x), np.max(x), len(x)*100)

# Calculating the polynomial from the coefficients
newy = 0
for i in range(len(coeffs)):
    newy = newy + coeffs[i]*newx**((len(x)-1)-i)

plt.scatter(x,y)
plt.plot(newx, newy)
```

```python
import numpy as np
from numpy import linalg as la
import math as m
import matplotlib.pyplot as plt
from scipy import interpolate as ip

x = np.array([1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9])
y = np.array([0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0])

# Applying the cubic intrapolation
f = ip.interp1d(x, y, kind = 'cubic')
newx = np.linspace(1.2,9.9,100) # defining x in range of the x array
newy = f(newx) # applying function f to the array newx to determine the corresponding y values

fig, ax = plt.subplots()
ax.scatter(x,y)
ax.plot(newx, newy)
plt.show()

# The fit is better, its way smoother than the polynomial

# Creating a new class : intrapolation
class intrapolation:
    def __init__(self, x, y): # initiation, class needs input of x and y 
        self.x = x
        self.y = y

    def spline(self): # spline intrapolation, same as above
        f = ip.interp1d(self.x, self.y, kind = 'cubic')
        newx = np.linspace(np.min(self.x), np.max(self.x), len(self.x)*100) # len(self.x*100) makes sure there's enough data points no matter the length of the array
        newy = f(newx)

        fig, ax = plt.subplots()
        ax.plot(newx, newy)
        plt.show()

    def polynomial(self): # polynomial intrapolation, same as previous exercise
        a_val = []
        for i in reversed(range(len(self.x))):
            a = self.x**(i)
            a_val.append(a)
        
        a_val_array = np.array(a_val)
        matrix = a_val_array.reshape(len(self.x),len(self.x))
        A = matrix.T
        y2 = self.y.reshape(len(self.y),1)
        
        coeffs = la.solve(A,y2)

        newx = np.linspace(np.min(self.x), np.max(self.x), len(self.x)*100)
        newy = 0
        for i in range(len(coeffs)):
            newy = newy + coeffs[i]*newx**((len(self.x)-1)-i)

        fig, ax = plt.subplots()
        ax.plot(newx, newy)
        plt.show()

    def linear(self): # linear intrapolation, same logic as spline but different "kind" in interp1d function
        f = ip.interp1d(self.x, self.y, kind = 'linear')
        newx = np.linspace(np.min(self.x), np.max(self.x), len(self.x)*100)
        newy = f(newx)
        
        fig, ax = plt.subplots()
        ax.plot(newx, newy)
        plt.show()
        
    def X_hat(self, X, interp):
        if np.min(self.x) <= X <= np.max(self.x): # checking if X is within range of x
            if interp == 'spline':
                f = ip.interp1d(self.x, self.y, kind = 'cubic') # calcualtes the intrapolation function based on x
                y = f(X) # applies the function to X (user input) and returns y
                return y
            elif interp == 'polynomial':
                a_val = []
                for i in reversed(range(len(self.x))):
                    a = x**(i)
                    a_val.append(a)
        
                a_val_array = np.array(a_val)
                matrix = a_val_array.reshape(len(self.x),len(self.x))
                A = matrix.T
                y2 = self.y.reshape(len(self.y),1)
                
                coeffs = la.solve(A, y2)

                newx = np.linspace(np.min(self.x), np.max(self.x), len(self.x)*100)
                newy = 0
                for i in range(len(coeffs)):
                    newy = newy + coeffs[i]*X**((len(self.x)-1)-i) # instead of calulation polynomials for an array of values of x it calulates the value of the polynomial at X
                
                return newy

            elif interp == 'linear':
                f = ip.interp1d(self.x, self.y, kind = 'linear') # same as spline
                y = f(X)
                return y
            else:
                print("Not a valid intrapolation method") # shows error if user inputs a not valid intrapolation method
        else:
            print("X not within range") # shows error if user input an X outside the range of x

X = float(input("X:"))
interp = input("Intrapolation:")

U = intrapolation(x = np.array([1.2, 2.0, 3.1, 4.4, 4.9, 5.8, 6.9, 8.0, 9.1, 9.9]), y = np.array([0.3, 0.5, 0.7, 0.8, 1.3, 1.6, 2.0, 2.3, 2.9, 3.0]))

U.spline()
U.polynomial()
U.linear()

print(U.X_hat(X, interp))
```

```python
import numpy as np

def bilinearIP(I, col, row): # Make it X, Y compatible
    '''I is a two dimensional array and iF,jF is the position at
       which we want to interpolate. 'iF' must be bigger or equal to 0 and
       smaller than the highest index of I in i. The same applies for jF'''
    rowlen, collen= I.shape
    i = int(row)
    j = int(col)
    if row - i == 0 and col - j == 0:  # Falls exactly on a gridpoint
        return I[i,j]

    inbounds = (0 <= j <= collen-1) and (0 <= i <= rowlen-1)
    if not inbounds:
        Iintp = None
    else:
        mu =  row - i
        Lambda = col - j
        # If we are at the last pixel at the right or bottom, then there
        # is not a neighbour, so we take the pixel itself as the neighbour
        I0 = I[i,j]
        I1 = I[i,min(j+1,collen-1)]
        I2 = I[min(i+1,rowlen-1),j]
        I3 = I[min(i+1,rowlen-1),min(j+1,collen-1)]
        Iintp = (1-Lambda)*(1-mu)*I0 + Lambda*(1-mu)*I1 + mu*(1-Lambda)*I2 + Lambda*mu*I3
    return(Iintp)

I = np.array([[2.0,4.1], [2.5,3.8]])
print(I)

row_y = 0.7 # Row position is y value
col_x = 0.2  # Column position is x value

Iintp = bilinearIP(I, col_x, row_y)
if Iintp:
    print(f"The interpolated intensity at i,j={row_y},{col_x} => x, y={col_x}{row_y} is {Iintp:.6f}")
else:
    print("One of the indices was outside the limits")
```
