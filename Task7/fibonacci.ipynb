```python
class Fibonacci:
    def __init__(self, n, m):
        self.n = n
        self.m = m
    def fibo_n(self):
        a = 0
        b = 1
        for i in range(2, self.n):
            c = a + b
            a = b
            b = c
        return b
    def fibo_list(self):
        fiblist = []
        a = 0
        b = 1
        l = 1
        while l < self.n:
            if a > self.m and a%self.m == 0:
                fiblist.append(a)
            c = a + b
            a = b
            b = c
            l = l + 1
        return fiblist
                
fib = Fibonacci(100,7)

print(fib.fibo_n())
print(fib.fibo_list())
        
```
