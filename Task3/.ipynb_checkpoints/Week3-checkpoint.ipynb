```python
# Question 3.1

import math

# Masses of the plants in the Solar System in order: Sun’, ‘Jupiter’, ‘Saturn’, ‘Neptune’, ‘Uranus’, ‘Earth’, ‘Venus’, ‘Mars’, ‘Mercury’, ‘Moon’, ‘Pluto’.

masses = [1.9891e+30, 1.8986e+27, 
          5.6846e+26, 1.0243e+26, 8.6810e+25,
          5.9736e+24, 4.8685e+24, 6.4185e+23, 
          3.3022e+23, 7.349e+22, 1.25e22]

# Finding masses larger than the mass of the Moon

newlist = []
for m in masses:
    if m > 7.349e+22:
        newlist.append(m)
        
print(f"Masses larger then the mass of the Moon: {newlist}")


# Finding the average of the last five masses from the original list

list2 = masses[6:]
print(f"The new list is {list2}.")

totalmass = sum(list2)
N = len(list2)
averagemass = totalmass / N

print(f"The average mass of the elements in the new list is {averagemass}.")
```
