```python
#Question 2.12

from math import atan2, degrees

x = float(input('What is the x value: '))
y = float(input('What is the y value: '))

angle = atan2(y, x)
ad = degrees(angle)  #Angle in degrees

if 0 < ad < 90 :
    print(f"The point {x, y} belongs to the first quadrant and it makes the angle {ad} with the x axis")
elif 90 < ad < 180 :
    print(f"The point {x, y} belongs to the second quadrant and it makes the angle {ad} with the x axis")
elif -180 < ad < -90 :
    print(f"The point {x, y} belongs to the third quadrant and it makes the angle {ad} with the x axis")
else :
    print(f"The point {x, y} belongs to the fourth quadrant and it makes the angle {ad} with the x axis")
```
