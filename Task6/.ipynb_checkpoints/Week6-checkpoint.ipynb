```python
# Question 1

import numpy as np
from astropy.io import fits
from scipy import stats
import matplotlib.pyplot as plt

# 1. Opening the file
hdulist = fits.open('m101.fits')

# 2. Lookign at the info
print(hdulist.info())

# Nr of HDUs: 2, image at PRIMARY (index 0), number type of the data: int16 - it can handle numbers from -32768 to 32768

# 3. Showing the headers and printing the first 5
hdr = hdulist[0].header

for key in hdr[:5]:
    print(f"{key:18s}= {hdr[key]}")

# yes the data type does correspond ( BITPIX 16 )

# 4. Read the data in a variable called dat.

dat = hdulist[0].data
a = np.shape(dat)
print(f"Shape: {a}") # It corresponds to the info, it correesponds to NAXIS1 but not NAXIS
print(f"Data type: {dat.dtype}") # gives >i2, signed 2-byte int, corresponds to stuff above bc 2 bytes = 16 bit

# 5. Find the image data: minimum, maximum, mean, standard deviation, median, mode.

dt = dat.flatten()
min, max, mean, std, median, mode = np.min(dt), np.max(dt), np.mean(dt), np.std(dt), np.median(dt), stats.mode(dt)[0]
print(f"min = {min}, max = {max}, mean = {mean}, std = {std}, mediad = {median}, mode = {mode}")

# 6. Plotting histogram

fig, ax = plt.subplots()

ax.hist(dt)
ax.axvline(mean, color = 'red', label = 'mean') 
ax.axvline(median, color = 'green', label = 'median') 
ax.axvline(mode, color = 'yellow', label = 'mode') 
ax.legend()

# 7. Converting to floating points and calculating the median

data = np.asarray(dat)
median2 = np.median(data)

# Finding the values assignined to the keywords and plotting the image

CTYPE1 = hdr['CTYPE1']
CTYPE2 = hdr['CTYPE2']

fig2, ax2 = plt.subplots()
bruh = ax2.imshow(data, vmin = median2)
ax.set_xlabel(CTYPE1)
ax.set_ylabel(CTYPE2)
fig2.colorbar(bruh, label = 'Counts')

# 8. Adding BUNIT

hdr['BUNIT'] = 'Counts'

hdulistout = fits.PrimaryHDU(data)
hdulistout.header = hdr    # We can copy the old header
hdulistout.writeto('newdata.fits', overwrite=True, output_verify='ignore')
output_verify='ignore'
```

```python
# Question 3

import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import math

hdulist = fits.open('m101.fits')
data = hdulist[0].data
fig, ax = plt.subplots()

# Applying a mask to see where the noise is.

mask = np.percentile(data, 40)
ax.imshow(data, vmax = mask)

# taking a slice of the data

area = np.array(data[0:20, 0:20])
noise = area.astype(float)

rms = math.sqrt((1/400)*(np.sum(noise**2)))

print(f"rms = {rms}, std = {np.std(noise)}, mean = {np.mean(noise)}")
```

```python
# Question 5

import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import math

m0 = fits.open('noisemap0.fits')
map0 = m0[0].data
m1 = fits.open('noisemap1.fits')
map1 = m1[0].data
m2 = fits.open('noisemap2.fits')
map2 = m2[0].data

# Calculating stds
std0, std1, std2 = np.std(map0), np.std(map1), np.std(map2)

# stacking data
cube = np.stack((map0, map1, map2), axis = 0)

mean = np.mean(cube, axis = 0) # mean image
std = np.std(mean) # std of the pixels in the mean image

std_ = (1/3)*math.sqrt(std0**2+std1**2+std2**2)

print(f"standard diviation of the mean image: {std}, calculated standard deviation: {std_}")
```

```python
# Question 6

import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import math
from astropy.table import Column, Table

# Open the FITS file
hdul = fits.open('APOGEE_stars (1).fits')
print(hdul.info()) 
# There is no difference in reading the table using astropy.io
# The difererence in structure: the type, dimensions and format are dofferent, the data is in the first extension (2nd header) 

# Access the primary data (in this case, it is a Table)
data = hdul[1].data  # Often, the first extension contains the table
colnames = hdul[1].columns.names # List of the column names

# Adding a DATAFLAG

mask = (data['STARFLAG'] == 0) & (data['ASPCAPFLAG'] == 0) & (data['SNR'] > 10) & (data['GAIAEDR3_PARALLAX'] / data['GAIAEDR3_PARALLAX_ERROR'] > 5) & (~np.isnan(data['GAIAEDR3_PHOT_G_MEAN_MAG'])) & (~np.isnan(data['GAIAEDR3_PHOT_BP_MEAN_MAG'])) & (~np.isnan(data['GAIAEDR3_PHOT_RP_MEAN_MAG']))

newdata = []

for i in range(len(data)): # looping through the rows analysing whther the values in the columns satisfy the conditions
    if mask[i] == True:
        newdata.append(1)
    else:
        newdata.append(0)

data = Table(data, names = colnames)
data.add_column(newdata, name = 'DATAFLAG', index = -1)

# Calculating the absolute magnitude

mag = data['GAIAEDR3_PHOT_G_MEAN_MAG'] + 5*np. log10(data['GAIAEDR3_PARALLAX']/1000) + 5

# Creating a new mask and creating values for the new column 

mask2 = (data['DATAFLAG'] == 0)

newlist = []

for i in range(len(data)):
    if mask[i] == True:
        newlist.append(-1)
    else:
        newlist.append(mag[i])

data.add_column(newlist, name = 'Absolute magnitude') # adding the new column

data.write('APOGEE_update.fits', overwrite=True)
```

```python
# 6.14 : Update your fibonacci code using OOP

class Fibonacci:
    def __init__(self, n, m):
        self.n = n
        self.m = m
    def fibo_n(self):
        a = 0
        b = 1
        for i in range(2, self.n):
            c = a + b
            a = b
            b = c
        return b
    def fibo_list(self):
        fiblist = []
        a = 0
        b = 1
        l = 1
        while l < self.n:
            if a > self.m and a%self.m == 0:
                fiblist.append(a)
            c = a + b
            a = b
            b = c
            l = l + 1
        return fiblist
                
fib = Fibonacci(100,7)

print(fib.fibo_n())
print(fib.fibo_list())
        
```
