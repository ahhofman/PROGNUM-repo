```python
import math

stars = {
    "Sirius": 9940,
    "Sun": 5800,
    "Proxima Centauri": 3042,
    "Beta Centauri" : 25000
}

radii = {
    "Sirius": 1.71,
    "Sun": 1,
    "Proxima Centauri": 0.154,
    "Beta Centauri": 8.4
}

planets = {
    "Sirius": [4.1, 8.2, 5.0],
    "Sun": [0.39, 0.72, 1.0, 1.80],
    "Proxima Centauri": [0.05, 0.12],
    "Beta Centauri": [100.2, 153.5, 250.3, 280.5]
}

T_sun = 5800

def spectral_class(temp): # function which determines the spectral type for each teperature
    if temp >= 30000:
        return "O"
    elif temp >= 10000:
        return "B"
    elif temp >= 7500:
        return "A"
    elif temp >= 6000:
        return "F"
    elif temp >= 5200:
        return "G"
    elif temp >= 3700:
        return "K"
    else:
        return "M"

for star in stars:
    if star in radii:
        T = stars[star] # temperature taken from the first dictionary
        R = radii[star] # radii taken from the second dictionary

        # printing the spectral class
        s_class = spectral_class(T)
        print(f"Star: {star} is {s_class}-type star")

        # solar luminosity
        L = R**2 * (T / T_sun)**4

        # habitable distance
        d = L**0.5

        # habitable zone edges
        inner = math.sqrt(d**2/1.06)
        outer = math.sqrt(d**2/0.36)

        # checking which planets are habitable
        for planet_distance in planets[star]:
            if inner <= planet_distance <= outer:
                result = "Habitable"
            else:
                result = "Not Habitable"

            print(f"Planet at {planet_distance} AU: {result}")

        print() # break line so its more visable
```

```python
import math

stars = {
    "Sirius": {"radius": 9940, "temp": 9940, "type": "A-type"},
     "Sun": {"radius": 1.0, "temp": 5778, "type": "G-type"},
    "Proxima Centauri": {"radius": 0.154, "temp": 3042, "type": "M-type"},
    "Beta Centauri": {"radius": 8.4, "temp": 25000, "type": "B-type"}
}
planets = {
    "Sirius": [4.1, 8.2, 5.0],
    "Sun": [0.39, 0.72, 1.0, 1.80],
    "Proxima Centauri": [0.05, 0.12],
    "Beta Centauri": [100.2, 153.5, 250.3, 280.5]
}

T_Sun = 5778

for star, info in stars.items():
    # printing the spectral class
    print(f"{star}, is a {info["type"]} star")

    # calculating the solar luminosity
    L = info["radius"]*((info["temp"]/T_Sun)**4)

    # calculating the habitable zone
    d = math.sqrt(L)

    # calculating the edges
    inner = math.sqrt((d**2)/1.06)
    outer = math.sqrt((d**2)/0.36)

    # checking if planets are within habitable zone
    for distance in planets[star]:
        if inner < distance < outer:
            print(f"Planet at {distance} AU : Habitable")
        else:
            print(f"Planet at {distance} AU : Not habitable")
    print()

    
```

```python
# Generating the first 60 prime numbers

def get_n_primes(n):
    """Generates a list of the first n prime numbers."""
    primes = []
    num = 2
    while len(primes) < n: # keeps the loop running until there is n prime numbers in the list
        for i in range(2, int(num**0.5) + 1): # the range of i is between 2 and the closest integer to sqrt of num
            if num % i == 0: # if num / i is an integer it breakes the loop
                break
        else:
            primes.append(num)
        num += 1
    return primes

prime_list = get_n_primes(60)

# Identify the 15th, 37th, and 60th primes
# Note: Human 15th = Index 14
dir_1 = prime_list[14]
dir_2 = prime_list[36]
dir_3 = prime_list[59]

# Construct the hint path

print(f"The 15th prime is: {dir_1}")
print(f"The 37th prime is: {dir_2}")
print(f"The 60th prime is: {dir_3}")
print(f"Constructed Path Hint: /{dir_1}/{dir_2}/{dir_3}/DisableCode.py")
```

```python
# Finding the first 24 fibo numbers

a = 1
b = 2
fibnumbers = [1]

for i in range(23):
    c = a + b
    a = b
    b = c
    fibnumbers.append(a)
print(fibnumbers)

# Even fib numbers from index 2

selected_fibs = fibnumbers[2::2]
print(selected_fibs)

# 3. Use the selected numbers as indices to find the password
# Note: Since the Fibonacci numbers grow very large quickly, we use 
# modulo (%) to ensure the index stays within the psw_reminder length.
psw_reminder = "abcdefghijklmnopqrstuvwxyz0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*()"

password_chars = []
for index in selected_fibs:
    char_index = index % len(psw_reminder)
    password_chars.append(psw_reminder[char_index])

password = "".join(password_chars)

print(f"Selected Fibonacci Numbers: {selected_fibs}")
print(f"Derived Password: {password}")
```
