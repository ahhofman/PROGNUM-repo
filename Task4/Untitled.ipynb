```python
# Question 4.21

import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
from matplotlib.pyplot import subplots, show
from scipy.stats import multivariate_normal

X = np.random.normal(loc=0, scale=1, size=100000)
Y = np.random.normal(loc=2, scale=1, size=100000)

x = np.linspace(-6, 8, 1000)

# Plotting the histogram

fig, frame = subplots(1, 1)
frame.hist(X, bins=30, alpha=0.5, label='X', edgecolor='black', density=True)
frame.hist(Y, bins=30, color='r', alpha=0.5, label='Y', edgecolor='black', density=True)

frame.plot(x, norm.pdf(x, 0, 1), color='black', label='X(PDF)', linestyle = ':')
frame.plot(x, norm.pdf(x, 2, 1), color='black', label='Y(PDF)', linestyle = '--')
plt.legend(loc='upper right')
plt.grid(axis='y', linestyle = '--')
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.title("1D Histogram + PDF")
plt.xlim(-6,8)
plt.show()

# 2D hist

plt.hist2d(X, Y, cmap='Blues',  bins=30)

mean = [0, 2]
cov = [[1, 0],
       [0, 1]]

Xg, Yg = np.meshgrid(X, Y)

rv = multivariate_normal(mean, cov)
Z = rv.pdf(np.dstack((Xg, Yg)))

levels = []
for k in [1, 2, 3]:
    level = (1/(2*np.pi)) * np.exp(-k**2 / 2)
    levels.append(level)

plt.contour(Xg, Yg, Z, levels=levels, colors='red')
plt.clabel(plt.contour(Xg, Yg, Z, levels=levels), inline=True)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("2D Histogram + Contour")
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.grid(linestyle = '--')

plt.show()
```
